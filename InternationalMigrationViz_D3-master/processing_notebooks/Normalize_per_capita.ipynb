```python
import numpy as np
import pandas as pd
```

```python
# get the list of countries that Datamaps uses
countries  = pd.read_csv('../raw_data/CountriesList.txt', delimiter=' , ', index_col='Country Code')
countries
```

```python
label_cols = ['Country Name', 'Country Code']
year_cols = ['1990', '1995', '2000', '2005', '2010', '2015']
keep_cols = label_cols + year_cols

pop = pd.read_csv('../raw_data/Population.csv',
                  index_col='Country Code', usecols=keep_cols)

print("In datamaps but not populations:\n", countries.index.difference(pop.index), end='\n\n')
print("In populations but not datamaps:\n", pop.index.difference(countries.index), end='\n\n')
```

```python
gdp = pd.read_csv('../raw_data/GDP.csv',
                  index_col='Country Code', usecols=keep_cols)

print("In datamaps but not gdp:\n", countries.index.difference(pop.index), end='\n\n')
```

```python
# we just want the rows that appear in the datamaps countries
pop = pop.reindex(pop.index & countries.index)
gdp = gdp.reindex(pop.index & countries.index)
print(all(pop.index == gdp.index))
```

```python
# ISO data for converting numeric UN codes to the alpha-3 codes used above
iso_codes = pd.read_csv('../raw_data/ISO_codes.csv', index_col='country-code')
def num_to_alpha3(num_code): 
    return iso_codes.loc[num_code, 'alpha-3']
iso_codes
```

```python
# Check out any countries with null entries from pop
pop[pop.isnull().any(axis=1)]
```

```python
# Check out any countries with null entries from GDP
gdp[gdp.isnull().any(axis=1)]
```

# Write GDP per capita data to processed_data

```python
gdp_per_cap = gdp.copy()
for year in year_cols:
    gdp_per_cap[year] = gdp[year] / pop[year]

gdp_per_cap.rename(columns={'Country Name': 'name'}, inplace=True)
gdp_per_cap.fillna(0).to_csv('../processed_data/GDP_per_capita.csv', index_label='id')
```

# Process UN migration data from .xlsx files

```python
bad_codes = [830]

migrate_1990 = pd.read_excel('../raw_data/Migrate_1990.xlsx', header=0)
migrate_1990 = migrate_1990.query('Numeric < 900 and Numeric != 830')
migrate_1990['Country Code'] = migrate_1990['Numeric'].apply(num_to_alpha3)
migrate_1990.set_index('Country Code', inplace=True)
migrate_1990 = migrate_1990.reindex(migrate_1990.index & pop.index)

good_countries = set(migrate_1990['Destination'])

def UN_name_to_alpha3(name):
    return migrate_1990.index[migrate_1990['Destination'] == name][0]

for ccol in migrate_1990.columns[6:]:
    if ccol not in good_countries:
        migrate_1990.drop(ccol, axis=1, inplace=True)
        print(ccol)
    else:
        migrate_1990.rename(columns={ccol: UN_name_to_alpha3(ccol)}, inplace=True)

migrate_1990.fillna(0.0)
```

```python
def get_migrate_df(year):
    df = pd.read_excel('../raw_data/Migrate_'+year+'.xlsx', header=0)
    
    df = df.query('Numeric < 900 and Numeric != 830')
    df['Country Code'] = df['Numeric'].apply(num_to_alpha3)
    df.set_index('Country Code', inplace=True)
    df = df.reindex(df.index & pop[year].dropna().index)
    
    good_countries = set(df['Destination'])
    
    for ccol in df.columns[6:]:
        if ccol not in good_countries:
            df.drop(ccol, axis=1, inplace=True)
        else:
            df.rename(columns={ccol: df.index[df['Destination'] == ccol][0]}, inplace=True)
    
    return df.fillna(0.0)

migrate = {}
for year in year_cols:
    print("Getting df for: ", year)
    migrate[year] = get_migrate_df(year)
    print(migrate[year].shape)
```

# Write Total Migrant Stock (as frac of population) to processed data

```python
total_stock = pd.DataFrame(migrate_1990['Destination'].copy())
total_stock.rename(columns={'Destination': 'name'}, inplace=True)

for year in year_cols:
    total_stock[year] = migrate[year]['Total'] / pop[year]

total_stock.fillna(0.0).to_csv('../processed_data/Total_stock_as_fraction.csv', index_label='id')
```
