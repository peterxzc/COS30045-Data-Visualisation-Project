```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
gdp = pd.read_csv('../processed_data/GDP_per_capita.csv', index_col='id')
stock = pd.read_csv('../processed_data/Total_stock_as_fraction.csv', index_col='id')
stock
```

```python
years = ['1990','1995','2000','2005','2010']

for year in years:
    print('YEAR: ', year)
    print('GDP (min, max): ', '(%f, %f)' % (gdp[year].min(), gdp[year].max()))
    gdp[year].hist()
    plt.show()
```

```python
for year in years:
    print('YEAR: ', year)
    print('Stock (min, max): ', '(%f, %f)' % (stock[year].min(), stock[year].max()) )
    stock[year].hist()
    plt.show()
```
